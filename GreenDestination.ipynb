```python
# Using the map function to convert column to boolean
data['Attrition']=data['Attrition'].map({'Yes':1,'No':0})
```

```python
# The attrition rate is the percentage of employees who have left the company over a certain period of time.
# Attrition Rate = ( Number of employees who left/ Total number of employees) ×100
```

```python
emp_left=(data['Attrition']==1)
```

```python
emp_left.sum()
```

```python
tot_emp=len(data)
```

```python
tot_emp
```

```python
# Attrition Rate = ( Number of employees who left/ Total number of employees) ×100
attrition_rate=(emp_left.sum()/tot_emp)*100
```

```python
# df=attrition_rate.sum()
```

```python
print('Attrition Rate:',attrition_rate)
```
